```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler, MinMaxScaler
```

```python
data = {"A": [1, 2, 3, 4, 5, 6, 7,8,12,12,20,6], "B": [10, 20, 30, 40, 50,100,20,40,50,200,60,55]}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df = pd.DataFrame(data)
```

```python
scaler = StandardScaler()

# fit and transform the data
scaled_data = scaler.fit_transform(df)

# create a new DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("Raw Data")
print(df)
print("\nScaled Data")
print(scaled_df)
```

```python
fig, ax = plt.subplots(1, 2)
sns.distplot(df["A"], ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_df["A"], ax=ax[1])
ax[1].set_title("Scale data")
```

```python
scaler = MinMaxScaler()

# fit and transform the data
normalized_data = scaler.fit_transform(df)

# create a new DataFrame with the normalized data
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)
print("Raw Data")
print(df)
print("\nNormalized Data")
print(normalized_df)
```

```python
fig, ax = plt.subplots(1, 2)
sns.distplot(df["A"], ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_df["A"], ax=ax[1])
ax[1].set_title("Normalization data")
```
